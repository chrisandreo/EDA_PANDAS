# What is the best-selling category? (BI Status = Net)

```python
pd.options.display.float_format = '{:.0f}'.format
df_best_selling = df[["category_name","priceTotal","BI Status"]]
df_best_selling["category_name"].value_counts()
```

```python
df_net = df_best_selling[df_best_selling["BI Status"]== "Net"]
df_net
```

```python
#let's drop \N ...
```

```python
df_net = df_net[~(df_net["category_name"] == r'\N')]
df_net["category_name"].value_counts()
```

```python
df_net = df_net.groupby("category_name").sum().sort_values(by = "priceTotal",ascending=False).reset_index()
df_net
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,5))
sns.barplot(y='category_name',x = "priceTotal",data = df_net)

None
```

```python
df["payment_method"].value_counts()
```

```python
df["Cod_or_Card"] = df["payment_method"] == "cod"
```

```python
df["Cod_or_Card"].replace(True,'Cod',inplace=True)
df["Cod_or_Card"].replace(False,'Card',inplace=True)
```

```python
df
```

```python
df['status'].value_counts()
```

```python
df[df['status'] == "cod"]
```

```python
df_Cod_or_Card_Cancelled = df[["Cod_or_Card","status","priceTotal"]]
df_Cod_or_Card_Cancelled
```

```python
df_Cod_or_Card_Cancelled = df_Cod_or_Card_Cancelled[(df_Cod_or_Card_Cancelled["status"]=='complete') | (df_Cod_or_Card_Cancelled["status"]=='canceled')]
df_Cod_or_Card_Cancelled
```

```python
df_Cod_or_Card_Cancelled2 = df_Cod_or_Card_Cancelled.groupby(["Cod_or_Card","status"]).count().reset_index()
df_Cod_or_Card_Cancelled2
```

# General percentage of status(canceled or complete)

```python
df_Cod_or_Card_Cancelled2['Total_Pct'] = (df_Cod_or_Card_Cancelled2["priceTotal"] / df_Cod_or_Card_Cancelled2["priceTotal"].sum()) *100
df_Cod_or_Card_Cancelled2
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Cod_or_Card',y = "Total_Pct",hue = "status",data = df_Cod_or_Card_Cancelled2,palette="Paired")
plt.ylim(0, 100)
plt.title("Order Status (percentage)",fontsize = 20)
None
```

# Card percentage of status(canceled or complete)

```python
dfCard=df_Cod_or_Card_Cancelled2[df_Cod_or_Card_Cancelled2['Cod_or_Card']=='Card']
dfCard
```

```python
dfCard['Total_Pct']=(dfCard['priceTotal']/dfCard['priceTotal'].sum()) *100
dfCard
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Cod_or_Card',y = "Total_Pct",hue = "status",data = dfCard,palette="Paired")
plt.ylim(0, 100)
plt.title("Order Status (percentage) - Cards",fontsize = 20)
None
```

# Cod percentage of status(canceled or complete)

```python
dfCod=df_Cod_or_Card_Cancelled2[df_Cod_or_Card_Cancelled2['Cod_or_Card']=='Cod']
dfCod
```

```python
dfCod['Total_Pct']=(dfCod['priceTotal']/dfCod['priceTotal'].sum()) *100
dfCod
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Cod_or_Card',y = "Total_Pct",hue = "status",data = dfCod,palette="Paired")
plt.ylim(0, 100)
plt.title("Order Status (percentage) - Cash on Delivery",fontsize = 20)
None
```
